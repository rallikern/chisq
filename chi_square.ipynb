```python
# f-spin mixing in a two-state mixing scenario
import sympy as sy

def solalpha(m1mf,m1sm,m1fs):
    m1ms = m1mf+m1sm-m1fs
    x = sy.Symbol('x')
    mixed = 2/5*(5*m1fs)**.5*(5*m1ms)**.5
    alph = sy.solve(m1ms*x**2 + mixed*x*(1-x**2)**.5 + (1-x**2)*m1fs - m1mf,x)
    return alph

def ratio(m1mf, m1sm, m1fs):
    a = np.array(solalpha(m1mf, m1sm, m1fs))
    rat = np.abs((1-a**2)**.5/a)
    return rat

#print(ratio([0.26, .13], [0.055, 0], [0.04, 0.01]))
#print(ratio([0.122, .010], [0.058, .006], [0.011, .002]))

def Ems(Emf, Esf, m1mf, m1sm, m1fs):
    rat = ratio(m1mf, m1sm, m1fs)
    E = (Emf + rat**2 * Esf) / (1+rat**2)    
    return E, rat

def Vmix(Emf, Esf, m1mf, m1sm, m1fs, size=10):
    Ems_temp, rat =  Ems(Emf, Esf, m1mf, m1sm, m1fs)
    V = np.array(np.abs((Emf -Ems_temp))/rat, dtype=float)
    return V


```
